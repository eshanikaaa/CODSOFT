```python
def to_do_list():
    tasks = []

    while True:
        print("\n--- Your To-Do List ---")
        print("1. Add New Task")
        print("2. View All Tasks")
        print("3. Complete a Task")
        print("4. Exit")

        action = input("Choose an option: ")

        if action == '1':
            task_count = input("How many tasks would you like to add? ")

            if task_count.isdigit():
                task_count = int(task_count)
                for _ in range(task_count):
                    new_task = input("Enter the task: ")
                    tasks.append({"task": new_task, "completed": False})
                    print("Task successfully added!")
            else:
                print("Please enter a valid number.")

        elif action == '2':
            if tasks:
                print("\nCurrent Tasks:")
                for idx, task in enumerate(tasks):
                    status = "Completed" if task["completed"] else "Pending"
                    print(f"{idx + 1}. {task['task']} [{status}]")
            else:
                print("\nNo tasks available.")

        elif action == '3':
            if tasks:
                try:
                    task_num = int(input("Which task number did you finish? ")) - 1
                    if 0 <= task_num < len(tasks):
                        tasks[task_num]["completed"] = True
                        print(f"Task {task_num + 1} marked as completed!")
                    else:
                        print("Task number out of range.")
                except ValueError:
                    print("Invalid input. Please enter a valid number.")
            else:
                print("No tasks to mark as completed.")

        elif action == '4':
            print("Goodbye! Your to-do list session has ended.")
            break

        else:
            print("Invalid choice. Please choose a valid option.")

if __name__ == "__main__":
    to_do_list()
```
